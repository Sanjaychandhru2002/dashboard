```python
from jupyter_dash import JupyterDash
from jinja2 import escape
import dash
import dash_core_components as css
import dash_html_components as html
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Car_data.csv")
df
```

```python
df["company"].unique()
```

```python
options=['Ford', 'Tata', 'Skoda', 'Audi', 'Toyota', 'BMW']
New_df=df[df['company'].isin(options)]
New_df
```

```python
from jupyter_dash import JupyterDash
from jinja2 import escape
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
app=dash.Dash()

app.layout = html.Div([
    html.Div("Second-Hand Car Dataset",style= { "color": "red", "text-align": "center"} ),
    html.Div("Barplot", style= {"color" : "blue", "text-align":"center"}),
     
    html.Div([
        dcc.Graph(
        id="barplpot",
        figure=px.bar(New_df, x="company",y="Price",color="company")
        )],style={'margin-right':100}),

    html.Div("Histogram",style={"color": "purple","text-align":"center"}),
    html.Div([
        dcc.Graph(
        id="Histogram",
        figure=px.histogram(New_df, x="company", color="company")
        )
    ]),
    
    html.Div("Scatter plot",style={"color":"blue","text-align":"center"}),
    html.Div([
        dcc.Graph(
        id="Scatter plot",
        figure=px.scatter(New_df, x="kms_driven", y="Price", color="company")
        )
    ]),
    
   
    html.Div("Box plot",style={"color":"orange","text-align":"center"}),
    html.Div([
        dcc.Graph(
        id="Box plot",
        figure=px.box(New_df['Price'])
        )
    ]),
    
    html.Div([
        dcc.Graph(
        id="Mosaic plot",
        figure=px.imshow(New_df.pivot_table(index='company', columns='kms_driven', values='Price'),
                     color_continuous_scale='RdYlGn', title='Mosaic Plot')
        )
    ]),
    
])
                                                                                                        
   
```

```python
if __name__=='__main__':
    app.run_server(port=4050)
```
